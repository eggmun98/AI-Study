```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
import matplotlib  # matplotlib 모듈 자체를 임포트
import matplotlib.pyplot as plt
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
print(f"PyTorch version: {torch.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
```

```python
print(f"Matplotlib version: {plt.matplotlib.__version__}")
```

```python
# 테스트 데이터셋 준비
testset = torchvision.datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transforms.ToTensor()
)

# 테스트 데이터로더 생성
batch_size = 256  # 과제 요구사항에 맞게 설정
testloader = torch.utils.data.DataLoader(
    testset,
    batch_size=batch_size,
    shuffle=False  # 테스트 데이터는 섞지 않음
)
```

```python
# 분류 모델 구현
class Model(nn.Module):
    def __init__(self, input_dim, n_dim):
        super().__init__()

        self.layer1 = nn.Linear(input_dim, n_dim)
        self.layer2 = nn.Linear(n_dim, n_dim)
        self.layer3 = nn.Linear(n_dim, 10)  # 10개 클래스로 출력 변경

        self.act = nn.ReLU()

    def forward(self, x):
        x = torch.flatten(x, start_dim=1)
        x = self.act(self.layer1(x))
        x = self.act(self.layer2(x))
        x = self.layer3(x)  # 마지막 활성화 함수 제거

        return x
```

```python
# 정확도 계산 함수
def accuracy(model, dataloader):
    cnt = 0
    acc = 0

    for data in dataloader:
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)  # device 사용

        preds = model(inputs)
        preds = torch.argmax(preds, dim=-1)

        cnt += labels.shape[0]
        acc += (labels == preds).sum().item()

    return acc / cnt

# 정확도 그래프 그리는 함수
def plot_acc(train_accs, test_accs, label1='train', label2='test'):
    x = np.arange(len(train_accs))
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, train_accs, label=label1)
    plt.plot(x, test_accs, label=label2)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Training and Test Accuracy over Epochs')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# 모델 초기화
input_dim = 28 * 28 * 1  # MNIST 이미지 크기
n_dim = 1024
model = Model(input_dim, n_dim).to(device)

# 학습 설정
n_epochs = 100  # 과제 요구사항
lr = 0.001      # 과제 요구사항
loss_fn = nn.CrossEntropyLoss()  # CrossEntropyLoss 사용
optimizer = optim.SGD(model.parameters(), lr=lr)

# 정확도 기록용 리스트
train_accs = []
test_accs = []
```

```python
# 학습 루프
for epoch in range(n_epochs):
    total_loss = 0.
    
    # 학습 모드
    model.train()
    
    for data in trainloader:
        optimizer.zero_grad()  # 또는 model.zero_grad()
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)

        preds = model(inputs)
        loss = loss_fn(preds, labels)  # CrossEntropyLoss 사용
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    
    # 평가 모드
    model.eval()
    
    # 정확도 계산 및 저장
    train_acc = accuracy(model, trainloader)
    test_acc = accuracy(model, testloader)
    
    train_accs.append(train_acc)
    test_accs.append(test_acc)

    # 진행 상황 출력
    print(f"Epoch {epoch:3d} | Loss: {total_loss:.4f} | Train Acc: {train_acc:.4f} | Test Acc: {test_acc:.4f}")

# 학습 완료 후 정확도 그래프 그리기
plot_acc(train_accs, test_accs)
```
